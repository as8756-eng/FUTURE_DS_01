## Visualize Satisfaction Data

### Subtask:

```python
informative_counts = df['5. Was the event informative?'].value_counts()

# Create a bar chart for informative_counts
informative_counts.plot(kind='bar', color='lightcoral', figsize=(8, 5))

# Set title and labels
plt.title('Distribution of Informativeness Rating')
plt.xlabel('Response')
plt.ylabel('Count')

# Rotate x-axis labels if needed
plt.xticks(rotation=0)

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
